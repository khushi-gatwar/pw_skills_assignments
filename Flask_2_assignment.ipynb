# Q1. Explain GET and POST methods.

## GET Method:

- Purpose: The GET method is used to request data from a specified resource.
- Data in URL: Data is appended to the URL in the form of query parameters.
- Visibility: Query parameters are visible in the URL, making them suitable for simple data retrieval.
- Caching: GET requests can be cached by browsers, and they remain in the browser's history.

## POST Method:

- Purpose: The POST method is used to submit data to be processed to a specified resource.
- Data in Request Body: Data is sent in the request body rather than in the URL. This makes it suitable for sending large amounts of data or sensitive information.
- Visibility: The data is not visible in the URL, providing a more secure method for transmitting data.
- Not Cached: POST requests are not cached by browsers, and they do not remain in the browser's history.
- Request Length: POST requests have no length restrictions on the amount of data that can be sent.

# Q2. Why is request used in Flask?

request object is used to access incoming request data when processing HTTP requests. It allows developers to access form data, query parameters, files, and other information sent in the HTTP request.



# Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the client (browser) to a different URL. It is commonly used when a user submits a form, and the server needs to redirect them to another page after processing the form data.

# Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are used to generate dynamic HTML content by embedding Python code within HTML files. Flask uses the Jinja templating engine for this purpose. Templates allow the separation of logic (Python code) and presentation (HTML) in web applications.

The render_template() function in Flask is used to render templates by combining HTML templates with data provided by the server. It takes the name of the template file as an argument and can also include additional variables to pass data to the template.
